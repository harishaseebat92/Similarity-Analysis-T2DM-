```python
from Bio import SeqIO
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
from pylab import savefig
def eig_val(filename):
  xlst=[ ]
  ylst=[ ]
  zlst=[ ]
  dic = {'A': (0.9511, 0.309, -1), 
         'B': (0.9511, 0.309, -1), 
         'C': (0.809, 0.5878, -1), 
         'D': (-0.9511, -0.309, 1), 
         'E': (0.309, -0.9511, 1), 
         'F': (-0.309, 0.9511, 1), 
         'G': (0.5878, -0.809, -1), 
         'H': (0.809, -0.5878, 1), 
         'I': (-0.5878, 0.809, -1), 
         'J': (-0.5878, 0.809, -1), 
         'K': (-0.809, -0.5878, 1), 
         'L': (-0.809, 0.5878, -1), 
         'M': (0.5878, 0.809, 1), 
         'N': (-0.5878, -0.809, -1), 
         'O': (-0.5878, -0.809, -1), 
         'P': (0.309, 0.9511, -1), 
         'Q': (0.9511, -0.309, 1), 
         'R': (-0.309, -0.9511, 1), 
         'S': (0.0, -1.0, -1), 
         'T': (1.0, 0.0, -1), 
         'U': (1.0, 0.0, -1), 
         'V': (0.0, 1.0, -1), 
         'W': (-0.9511, 0.309, 1), 
         'X': (-0.9511, 0.309, 1), 
         'Y': (-1.0, 0.0, 1), 
         'Z': (-1.0, 0.0, 1)
         }
  for elem in filename:
    x, y, z = dic[elem]
    xlst.append(x)
    ylst.append(y)
    zlst.append(z)
  xcom=(np.mean(xlst))
  ycom=(np.mean(ylst))
  zcom=(np.mean(zlst))
  xlst = xlst-xcom
  ylst = ylst-ycom
  zlst = zlst-zcom

  ixx=np.sum(np.square(ylst)+np.square(zlst))
  iyy=np.sum(np.square(xlst)+np.square(zlst))
  izz=np.sum(np.square(xlst)+np.square(ylst))
  ixy=np.sum(np.multiply(xlst,ylst))
  iyx=ixy
  iyz=np.sum(np.multiply(ylst,zlst))
  izy=iyz
  ixz=np.sum(np.multiply(xlst,zlst))
  izx=ixz
  from numpy.linalg import eig
  evalue,evector=eig(np.array([[ixx,-ixy,-ixz],
                           [-iyx,iyy,-iyz],
                           [-izx,-izy,izz]]))
  return(evalue)
```
